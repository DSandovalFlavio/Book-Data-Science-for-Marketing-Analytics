```python
import pandas as pd

age_spend = pd.read_csv('age_spend.csv')
age_spend.head()
```

```python
from sklearn.model_selection import train_test_split

X = age_spend[['age']]
y = age_spend[['spend']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor

max2_tree_model = DecisionTreeRegressor(max_depth=2)
max2_tree_model.fit(X_train, y_train)

max5_tree_model = DecisionTreeRegressor(max_depth=5)
max5_tree_model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

predic_lineal = model.predict(X_test)
print(f'Model Lineal RMSE: {mean_squared_error(y_test, predic_lineal)**0.5}')

max2_tree_predic = max2_tree_model.predict(X_test)
print(f'Model max_depth=2 RMSE: {mean_squared_error(y_test, max2_tree_predic)**0.5}')

max5_tree_predic = max5_tree_model.predict(X_test)
print(f'Model max_depth=5 RMSE: {mean_squared_error(y_test, max5_tree_predic)**0.5}')
```

```python
import matplotlib.pyplot as plt

ages = pd.DataFrame({'age': range(18, 70)})

plt.scatter(X_test.age.tolist(), y_test.spend.to_list())
plt.plot(ages,model.predict(ages), color='red', linewidth=5)
plt.show()
```

```python
plt.scatter(X_test.age.tolist(), y_test.spend.to_list())
plt.plot(ages,max2_tree_model.predict(ages), color='g', linewidth=5)
plt.show()
```

```python
plt.scatter(X_test.age.tolist(), y_test.spend.to_list())
plt.plot(ages,max5_tree_model.predict(ages), color='k', linewidth=5)
```

```python
from sklearn.ensemble import RandomForestRegressor

max2_forest_model = RandomForestRegressor(max_depth=2, random_state=10)
max2_forest_model.fit(X_train, y_train)

max5_forest_model = RandomForestRegressor(max_depth=5, random_state=10)
max5_forest_model.fit(X_train, y_train)
```

```python
max2_forest_predic = max2_forest_model.predict(X_test)
print(f'Model RandomForest max_depth=2 RMSE: {mean_squared_error(y_test, max2_forest_predic)**0.5}')

max5_forest_predic = max5_forest_model.predict(X_test)
print(f'Model RandomForest max_depth=5 RMSE: {mean_squared_error(y_test, max5_forest_predic)**0.5}')
```
